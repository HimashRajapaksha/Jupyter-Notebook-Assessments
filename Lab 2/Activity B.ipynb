```python
import pandas as pd
from apyori import apriori
```

```python
mData = pd.read_csv("movie_dataset.csv", header=None)
```

```python
mData.head()
```

```python
mData.shape
```

```python
records = []
for i in range(1, 7501):
    records.append([str(mData.values[i, j]) for j in range(0, 20)])
```

```python
# Apply the Apriori algorithm
frequent_itemsets = apriori(records, min_support=0.01, min_confidence=0.0, min_lift=0.0, min_length=2)

# Convert the result into a list
frequent_itemsets_results = list(frequent_itemsets)

# Display the frequent itemsets
for itemset in frequent_itemsets_results:
    if len(itemset.items) == 2:
        print(f"Frequent Itemset: {itemset.items}, Support: {itemset.support}")
```

```python
# Apply the Apriori algorithm with specific min_confidence and min_lift
association_rules = apriori(records, min_support=0.01, min_confidence=0.5, min_lift=0.0, min_length=2)

# Convert the result into a list
association_results = list(association_rules)

# Display the association rules
for rule in association_results:
    for ordered_stat in rule.ordered_statistics:
        if len(ordered_stat.items_base) == 1 and len(ordered_stat.items_add) == 1:
            print(f"Rule: {ordered_stat.items_base} -> {ordered_stat.items_add}")
            print(f"Support: {rule.support}")
            print(f"Confidence: {ordered_stat.confidence}")
            print(f"Lift: {ordered_stat.lift}")
            print("---------------------------")
```
